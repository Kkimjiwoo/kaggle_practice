```python
import os
from sklearn.model_selection import train_test_split
```

```python
x = feature 
y = target
```

```python
# 학습데이터(test)와 평가데이터(train) 8:2 비율로 만들기

# train_x : 독립변수 x(특징 데이터)의 평가데이터
# test_x : 독립변수 x(특징 데이터)의 학습데이터
# train_y : 종속변수 y(타겟 데이터)의 평가데이터
# test_y : 종속변수 y(타겟 데이터)의 학습데이터
train_x, test_x, train_y, test_y = train_test_split(x,y, test_size =0.2, random_state=42 )
# x를 0.2의 비율을 train_x의 변수로 대입 
# x를 1-0.2의 비울로 test_x의 변수로 대입
# y도 동일한 방식

print(train_x.shape, test_x.shape, train_y.shape, test_y.shape) 
```

```python
# 기본적인 random forest 모형

from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score  # 정확도 함수

# clf : Classifier (분류)
clf = RandomForestClassifier(n_estimators=20, max_depth=5, random_state=20)
# n_estimators=20 20개의 의사결정 트리를 만든다 
#>> 이 값이 클수록 모델의 복잡도 증가, 성능 향상 증가
# max_depth=5 트리의 최대 깊이
#>> 모델의 깊이를 제한함으로써 모델의 과적합 방지
#>> 각 트리가 너무 깊게 성장하지 않도록 함

clf.fit(train_x, train_y)
# fit 메서드를 호출하면 입력데이터와 출력데이터 사이의 관계를 학습

predict1 = clf.predict(test_x) 
print(accuracy_score(test_y, predict1))
# predict1은 test_x에 대한 에측된 출력값을 계산
# accuracy_score 함수는 예측된 출력 값 predict1과 실제 정답 데이터 test_y를 
# 비교하여 분류 모델의 정확도를 계산

# 1에 가까울수록 모델이 더 정확하게 예측

# 약 61%의 정확도 
```

```python
# Random Forest Sample 개수 증가

clf = RandomForestClassifier(n_estimators=100, max_depth=20, random_state=20)
# 적합
clf.fit(train_x, train_y)
# 결과 예측
predict2 = clf.predict(test_x) 
print(accuracy_score(test_y, predict2))

# 약 78%의 정확도
```
